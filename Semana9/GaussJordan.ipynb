```python
def TriangularSuperior(A,b=None):

  A = np.array(A, dtype=float, copy=True)
  n = A.shape[0]
  y = None if b is None else np.array(b, dtype=float, copy=True)
  print(y)
  # k es la coordenada k
  for k in range(n):
    #print(k)
    if A[k,k] == 0:
      raise ValueError("La matriz problematica")
    else:
      pivote = A[k,k]
      for i in range(k+1,n):
        m_i_k = A[i,k]/pivote
        # Restamos a la fila j, el m_ik por la fija j
        A[i, k:] = A[i, k:] - m_i_k*A[k,k:]
        if y is not None:
          y[i] = y[i] - m_i_k*y[k]

  return (A,y) if b is not None else A
```

```python
U,y = TriangularSuperior(A,b)
```
