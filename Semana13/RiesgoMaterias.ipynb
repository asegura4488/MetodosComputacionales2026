```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.optimize import minimize
import itertools

df = pd.read_csv('Semana13/Data/datos_materias_riesgo.csv')
df
```

```python
# Umbral
nota_aprobacion = 3.0
peso_p1 = 0.30
peso_p2 = 0.30
peso_faltante = 0.40
```

```python
df['Nota_acumuladta'] = peso_p1 * df['Parcial_1'] + peso_p2 * df['Parcial_2']
df['Nota_necesaria'] = (nota_aprobacion-df['Nota_acumuladta']) / peso_faltante
df
```

```python
df['R_nota'] = df['Nota_necesaria'] / 5
# riesgo por el profesor
df['R_profesor'] = df['Dificultad_profesor_1a5'] / 5
# riesgo por horas
df['R_tiempo'] = df['Horas_estudio_semana'] / df['Horas_estudio_semana'].max()
df
```

```python
def riesgo_horario(hora):

  h = int(hora.split(':')[0])
  print(h)
  if h <= 7:
    return 0.7
  elif 10 <= h <= 14:
    return 0.2
  elif h >= 18:
    return 0.6
  else:
    return 0.4
```

```python
df['R_horario'] = df['Hora_clase'].apply(riesgo_horario)
df
```
