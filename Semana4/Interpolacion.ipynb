```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
X = np.array([1,2,3])
Y = np.array([0,2.5,4])
```

```python
#X = np.array([1,2])
#Y = np.array([5,6])
```

```python
plt.scatter(X,Y,color='r')
```

```python
# Programas la base
def Lagrange(x,X,i):

  L = 1 # Productoria
  for j in range(len(X)):
    if i != j:
      L *= (x-X[j])/(X[i]-X[j])
  return L
```

```python
# Definir un simbolo
x = sym.Symbol('x',real=True)
x**2
```

```python
L0 = Lagrange(x,X,1).simplify()
```

```python
L0
```

```python
def Interpolate(x,X,Y):

  Poly = 0
  for i in range(len(X)):
    Poly += Y[i]*Lagrange(x,X,i)
  return Poly
```

```python
p1 = Interpolate(x,X,Y).simplify()
p1
```

```python
p1 = sym.Poly(p1)
p1.all_coeffs()
```

```python
p1.all_coeffs()
```

```python
x1 = np.linspace(X[0],X[-1],100)
y1 = Interpolate(x1,X,Y)
```

```python
plt.scatter(X,Y,color='r')
plt.plot(x1,y1)
```
