```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
X = np.array([1,2,3,4,6])
Y = np.array([0,2.5,4,3,1])
```

```python
plt.scatter(X,Y,color='r')
```

```python
# Programas la base
def Lagrange(x,X,i):

  L = 1 # Productoria
  for j in range(len(X)):
    if i != j:
      L *= (x-X[j])/(X[i]-X[j])
  return L
```

```python
# Definir un simbolo
x = sym.Symbol('x',real=True)
x**2
```

```python
L0 = Lagrange(x,X,1).simplify()
```
