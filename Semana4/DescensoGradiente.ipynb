```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
import time
import copy
```

```python
def Plotter(e):

  fig = plt.figure(figsize=(8,3))
  ax = fig.add_subplot(121)
  ax1 = fig.add_subplot(122)
  ax.set_title('iteracion: {:.0f}'.format(e),fontsize=10)

  return ax,ax1
```

```python
ax,ax1 = Plotter(6)
```

```python
def f(x):
  return x**2 + 10*np.sin(x)
```

```python
xvalues = np.linspace(-10,10,50)
yvalues = f(xvalues)
```

```python
def Derivative(x,f,h=1e-4):
  return (f(x+h)-f(x-h))/(2*h)
```

```python
ax,ax1 = Plotter(6)
ax.plot(xvalues,yvalues)
ax
```

```python
def GradientDescent(f,epochs=100,lr=0.05):

  # Voy a elegir aleatoria la semilla del algoritmo
  #xn = np.random.uniform(-10,10)
  xn = -20

  for e in range(epochs):

    #El algoritmo
    xn1 = xn - lr*Derivative(xn,f)

    #Actualizamos el punt
    xn = xn1

    #Imprimimos

    clear_output(wait=True)
    ax,ax1 = Plotter(e)
    ax.plot(xvalues,yvalues)

    ax.scatter(xn,f(xn),color='r')

    plt.show()
```

```python
GradientDescent(f)
```
