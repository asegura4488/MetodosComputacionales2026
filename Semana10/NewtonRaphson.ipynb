```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Definamos el sistema usando una lista
Gfea = (lambda x,y,z: 3*x - np.cos(y*z) - 1.5, \
     lambda x,y,z: 4*x**2 - 625*y**2 + 2*y -1, \
     lambda x,y,z: np.exp(-x*y) + 20*z + 9.471975)
```

```python
Gbonita = (lambda x,y,z: 3*x - y - z - 1, \
     lambda x,y,z: -x + 3*y + z - 3, \
     lambda x,y,z: 2*x + y + 4*z - 7 )
```

```python
# Funcion de valor vectorial
Gfea[2](3,4,1)
```

```python
def GetVectorF(G,r):

  v = np.zeros(len(G))
  for i in range(v.shape[0]):
    v[i] = G[i](r[0],r[1],r[2])

  return v
```

```python
GetVectorF(Gfea,np.array([1,1,1]))
```

```python
# Calculemos la jacobiana
def GetJacobian(G,r,h=1e-4):

  J = np.zeros((r.shape[0],r.shape[0]))

  for i in range(r.shape[0]):
    J[i,0] = (G[i](r[0]+h,r[1],r[2]) - G[i](r[0]-h,r[1],r[2]))/(2*h)
    J[i,1] = (G[i](r[0],r[1]+h,r[2]) - G[i](r[0],r[1]-h,r[2]))/(2*h)
    J[i,2] = (G[i](r[0],r[1],r[2]+h) - G[i](r[0],r[1],r[2]-h))/(2*h)

  return J.T
```

```python
GetJacobian(Gfea,np.array([1,1,1]))
```

```python
def NewtonRaphson(G,r,tol=1e-10,maxiter=100):

  it = 0
  d = 10
  Distancias = np.array([])

  while d > tol:

    F = GetVectorF(G,r)
    J = GetJacobian(G,r)
    InvJ = np.linalg.inv(J)

    r_new = r - InvJ @ F

    d = np.linalg.norm(r_new - r)
    # mirando una distancia estilo cauchy
    r = r_new
    it += 1

    Distancias = np.append(Distancias,d)

    if it%50 == 0:
      print(d)

    if it > maxiter:
      break

  return r,Distancias
```

```python
r0 = np.array([1.,1.,1.])
r,Distancias = NewtonRaphson(Gfea,r0)
```
