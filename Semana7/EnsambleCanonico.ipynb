```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def SistemadeDosNiveles(eps=1, kb=1., T=1.7, N=10000):

  rng = np.random.default_rng(100)
  beta = 1/(kb*T)
  print(beta)

  # Definamos las energias
  E_levels = np.array([+eps,-eps], dtype=float)
  M_levels = np.array([+1,-1], dtype=float)

 # print(E_levels)
  w = np.exp(-beta*E_levels)
  #print(w)
  # Funcion particion
  Z = np.sum(w)
  #print(Z)
  # Calculemos la probabilidad
  p = w/Z
  #print(p)

  # Montecarlo
  # Muestreo
  c1 = p[0]
  u = rng.random(N)
  #print(u)
  idx = np.empty(N, dtype=int)
  idx[u < c1] = 0
  idx[u >= c1] = 1
  #print(idx)

  #plt.hist(idx)

  E = E_levels[idx]
  M = M_levels[idx]

  positiva = np.where(E == 1)
  negativa = np.where(E == -1)

  p_positiva = len(positiva[0])/N
  p_negativa = len(negativa[0])/N

  print(p_positiva,p_negativa)

  Mean_E = p_positiva*E_levels[0] + p_negativa*E_levels[1]

  print(E.mean(),Mean_E)
  #print(M.mean())
```

```python
SistemadeDosNiveles(T=0.7)
```
