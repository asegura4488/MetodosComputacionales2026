```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def SistemadeDosNiveles(eps=1, kb=1., T=1.7, N=10000):

  rng = np.random.default_rng(100)
  beta = 1/(kb*T)
  print(beta)

  # Definamos las energias
  E_levels = np.array([+eps,-eps], dtype=float)
  M_levels = np.array([+1,-1], dtype=float)

  print(E_levels)
  w = np.exp(-beta*E_levels)
  print(w)
  # Funcion particion
  Z = np.sum(w)
  print(Z)
  # Calculemos la probabilidad
  p = w/Z
  print(p)

  # Montecarlo
  # Muestreo
  c1 = p[0]
  u = rng.random(N)
  #print(u)
  idx = np.empty(N, dtype=int)
  idx[u < c1] = 0
  idx[u >= c1] = 1
  #print(idx)

  #plt.hist(idx)

  E = E_levels[idx]
  M = M_levels[idx]

  print(E.mean())
  print(M.mean())
```

```python
SistemadeDosNiveles(T=1000)
```
