```python
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# ----------------------------------
# Definición de vectores
# ----------------------------------
A = np.array([3, 2])
B = np.array([1, 4])

# ----------------------------------
# Operaciones vectoriales
# ----------------------------------
suma = A + B
resta = A - B

print("Vector A =", A)
print("Vector B =", B)
print("A + B =", suma)
print("A - B =", resta)

# ----------------------------------
# Crear figura
# ----------------------------------
plt.figure(figsize=(6,6))

plt.quiver(0,0,A[0],A[1],color='k',angles='xy',scale_units='xy',scale=1,label='A')
plt.quiver(0,0,B[0],B[1],color='b',angles='xy',scale_units='xy',scale=1,label='B')
plt.quiver(0,0,suma[0],suma[1],color='r',angles='xy',scale_units='xy',scale=1,label='A+B')
plt.quiver(0,0,resta[0],resta[1],color='purple',angles='xy',scale_units='xy',scale=1,label='A-B')

plt.xlim(-2,6)
plt.ylim(-3,7)

plt.axhline(0)
plt.axvline(0)

plt.xlabel("x")
plt.ylabel("y")
plt.title("Operaciones entre vectores")
plt.legend()

# ----------------------------------
# Guardar figura
# ----------------------------------
plt.savefig("Vectores_numpy.pdf",dpi=300,bbox_inches='tight')

plt.show()

# ----------------------------------
# Descargar automáticamente
# ----------------------------------
files.download("Vectores_numpy.pdf")
```
