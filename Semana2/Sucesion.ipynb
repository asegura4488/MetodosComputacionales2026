```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = lambda n: 1/n
```

```python
s(2)
```

```python
# Quiero evaluar la sucesion en varios puntos
N = 20
n = np.arange(1,N+1)
n
```

```python
a = np.zeros(N)
a
```

```python
for i in range(n.shape[0]):
  a[i] = s(i+1)
```

```python
a
```

```python
plt.plot(n,a,marker='o')
```
