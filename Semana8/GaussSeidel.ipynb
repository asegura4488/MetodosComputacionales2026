```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Longitud de la barra
L = 1
# Temperaturas
Tiz = 10.0
Tder = 100.0

N = 20

# Linspace
x = np.linspace(0,L,N)

h = x[1]-x[0]
print(h)
```

```python
# Definimos el mapa de temperatura
T = np.zeros(N)
T[0] = Tiz
T[-1] = Tder
T
```

```python
n_int = N - 2
```

```python
# Vamos a crear A
A = np.zeros((n_int,n_int))

for i in range(n_int):
  # Diagonal
  A[i,i] = -2.

  if i > 0:
    A[i, i-1] = 1.

  if i < n_int - 1:
    A[i, i+1] = 1.

A
```

```python
# Calculemos b
b = np.zeros(n_int)

b[0] = - Tiz
b[-1] = - Tder
b
```

```python
xb = np.linalg.solve(A,b)
xb
```

```python
Temperatura = np.zeros(N)
Temperatura[1:-1] = xb
Temperatura[0] = Tiz
Temperatura[-1] = Tder
Temperatura
```

```python
plt.scatter(x,Temperatura)
```
