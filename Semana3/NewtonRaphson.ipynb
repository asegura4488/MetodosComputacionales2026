```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Function(x):
  #return x**x - 100
  return 5*(1-np.exp(-x)) - x
```

```python
# Primer paso: pintamos la funcion
x = np.linspace(1, 5, 50)
y = Function(x)
```

```python
plt.plot(x, y)
plt.axhline(y=0, color='r', linestyle='--')
```

```python
# Como un curso de IP
for i in range(len(x)-1):
  # Necesito la funcion
  fi = Function(x[i])
  fj = Function(x[i+1])
  if fi*fj < 0:
    print(x[i],fi,fj)
    break
  #
```

```python
def Derivative(x,f,h=1e-6):
  return (f(x+h)-f(x-h))/(2*h)
```

```python
def GetNewtonMethod(f,df,xn,itmax=100,precision=1e-10):

  error = 1
  it = 0

  while error > precision and it < itmax:

    try:
      xn1 = xn - f(xn)/df(xn,f)
      print(xn,xn1)
      error = np.abs(xn1 - xn) # La nocion de cauchy
    except ZeroDivisionError:
      print("Division por cero")


    xn = xn1
    it += 1

  if it == itmax:
    return False
  else:
    return xn
```

```python
GetNewtonMethod(Function,Derivative,2.) # Este es la semilla del algoritmo
```
