```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Necesitamos definir el conjunto de soporte del campo
# Un lattice
N = 5
x = np.linspace(-5, 5, N)
y = np.linspace(-5, 5, N)
```

```python
x
```

```python
y
```

```python
# Dominio del soporte
X,Y = np.meshgrid(x,y)
```

```python
plt.scatter(X,Y)
```

```python
def f2(x,y):
  return np.sin(x) + 5*np.cos(y)
```

```python
# Definimos la imagen
Z = np.zeros_like(X)
Z
```

```python
for i in range(N):
  for j in range(N):
    Z[i,j] = f2(x[i],y[j])
```

```python
from matplotlib import cm
fig1 = plt.figure(figsize=(3,3))
ax1 = fig1.add_subplot(111, projection='3d')
ax1.plot_surface(X,Y,Z,cmap='viridis')
```

```python
# Para un campo vectorial necesitamos 2 componentes
Ex = np.zeros((N,N))
Ey = np.zeros((N,N))
```

```python
Ex
```

```python
#definir que es el campo
def GetField(x,y,r0):

  x0 = r0[0]
  y0 = r0[1]

  #print(x0,y0)
  # Ley de Coulomb
  Ex = ( x-x0 )/(  (x-x0)**2 + (y-y0)**2  )**(3/2)
  Ey = ( y-y0 )/(  (x-x0)**2 + (y-y0)**2  )**(3/2)

  return Ex,Ey
```

```python
fuentes = [-1,3.]
```

```python
GetField(1,3,fuentes)
```

```python
# Llenamos el valor del campo
for i in range(N):
  for j in range(N):
    Ex[i,j] = GetField(x[i],y[j],fuentes)[0]
    Ey[i,j] = GetField(x[i],y[j],fuentes)[1]
```

```python
plt.scatter(X,Y)
plt.scatter(*fuentes, color='r',s=200)

for i in range(N):
  for j in range(N):
    plt.quiver(x[i],y[j],Ex[i,j],Ey[i,j])
```
