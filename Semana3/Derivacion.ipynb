```python
import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-np.pi,np.pi,20)
x
```

```python
def funcion(x):
  return np.sin(x)
```

```python
y = funcion(x)
```

```python
plt.scatter(x,y)
```

```python
# Queremos definir la derivada derecha
def RightDerivative(x,f,h=0.001):

  d = (f(x+h) - f(x))/h

  return d
```

```python
RightDerivative(np.pi,funcion)
```

```python
derivada1 = RightDerivative(x,funcion)
```

```python
plt.plot(x,derivada1)
plt.plot(x,np.cos(x),label='Exacta')
plt.legend()
```
