```python
import numpy as np
import matplotlib.pyplot as plt
#from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-np.pi,np.pi,10)
x
```

```python
def funcion(x):
  return np.sin(x)
```

```python
y = funcion(x)
```

```python
plt.scatter(x,y)
```

```python
# Queremos definir la derivada derecha
def RightDerivative(x,f,h=0.1):

  d = (f(x+h) - f(x))/h

  return d
```

```python
# Queremos definir la derivada derecha
def LeftDerivative(x,f,h=0.1):

  d = (f(x) - f(x-h))/h

  return d
```

```python
# Queremos definir la derivada derecha
def CentralDerivative(x,f,h=0.1):

  d = (f(x+h) - f(x-h))/(2*h)

  return d
```

```python
derivada1 = RightDerivative(x,funcion)
derivada2 = LeftDerivative(x,funcion)
derivada3 = CentralDerivative(x,funcion)
```

```python
plt.scatter(x,np.cos(x),color='k',label='Exacta')
plt.plot(x,derivada1,label='Derecha')
plt.plot(x,derivada2,label='Izquierda')
plt.plot(x,derivada3,label='central')
plt.legend()
```

```python
# QUiero una derivada parcial
def f2(x,y):
  return x**2 + y**2
```

```python
def PartialDerivativeX(x,y,f2,h=0.001):

  d = (f2(x+h,y) - f2(x-h,y))/(2*h)

  return d
```

```python
PartialDerivativeX(1,2,f2)
```
