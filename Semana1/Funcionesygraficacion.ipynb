```python
# Existe una diferencia entre los parametros y el soporte
# introducir el concepto de funcion
def posicion(t,x0,v):
  # la fisica
  x = x0 + v*t
  return x
```

```python
posicion(3,1.2,7)
```

```python
# Vamos a generar el soporte
t = []
x = []
x1 = []
for i in range(10):
  t.append(i)
  x.append(posicion(i,1.2,7))
  x1.append(posicion(i,20,-3))
```
